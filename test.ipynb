```python
x = iris_pd["petal length (cm)"].values
y = (iris_pd["species"] == 0).astype(int).values
```

```python
def entropy(y):
    if len(y) == 0:
        return 0.0
    p1 = y.mean()
    p0 = 1 - p1

    e = 0.0
    if p0 > 0:
        e -= p0 * np.log2(p0)
    if p1 > 0:
        e -= p1 * np.log2(p1)
    
    return e
```

```python
thresholds = np.arange(0.0, 7.01, 0.01)
entropies = []
n = len(y)
for thr in thresholds:
    yL = y[x <= thr]
    yR = y[x > thr]

    HL = entropy(yL)
    HR = entropy(yR)

    wL = len(yL) / n
    wR = len(yR) / n

    entropies.append(wL * HL + wR * HR)

entropies = np.array(entropies)
```

```python
%matplotlib inline
plt.ioff()

fig, (ax_entropy, ax_scatter) = plt.subplots(
    2, 1,
    figsize = (12, 7.5),
    sharex = True,
    gridspec_kw = {"height_ratios": [1, 2.4]}
)
```

```python
ax_entropy.plot(
    thresholds,
    entropies,
    color = "black"
)

ax_entropy.set_ylabel("Entropy")
ax_entropy.set_title("Entropy vs Petal Length Threshold")
```

```python
sns.scatterplot(
    ax = ax_scatter,
    x = "petal length (cm)",
    y = "petal width (cm)",
    data = iris_pd,
    hue = "species",
    palette = "Set2",
)

ax_scatter.set_xlabel("petal length (cm)")
ax_scatter.set_ylabel("Petal width (cm)")

plt.tight_layout()
plt.show()
plt.ion()
```
